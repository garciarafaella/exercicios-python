# Ajustando nomes de produtos


Um programa que recebe um nome de produto e o padronize, deixando todas as letras minúsculas e removendo os espaços extras.

```python
produto = input("Digite o nome do produto: ")

produto_padronizado = " ".join(produto.strip().lower().split())

print(produto_padronizado)
```

# Decifrando pistas com palavras-chave

Um programa que solicia ao usuário uma palavra e exibe as três primeiras e as três últimas letras.

```python
palavra = input('Digite uma palavra: ')

palavra_primeiras = palavra[:3]
palavra_ultimas = palavra[-3:]

print(palavra_primeiras)
print(palavra_ultimas)
```

# Verificando o início e o fim de uma String

Está se desenvolvendo um sistema que verifica se os links de sites parceiros começam com https:// e terminam com .com. Esses critérios são obrigatórios para que o site seja aprovado no cadastro. Um programa que realiza essa validação de forma automática.

```python
link = input('Digite o link do site: ')

if link.startswith('https://') and link.endswith('.com'):
    print('Link aprovado!')

else:
  print('Link inválido!')
```

# Encontrando números em um texto

Um programa que recebe um texto com uma descrição e extrai um número diretamente e confirma se o texto está correto, sem a necessidade de trabalhar com listas ou loops, e exibe esse número.

```python
import re

texto = input("Digite: ")

resultado = re.search(r"\d+", texto)

if resultado:
    print("Número encontrado:", resultado.group())
else:
    print("Texto inválido: nenhum número encontrado.")
```

# Validando nomes com Regex

Um programa que solicite um nome ao usuário e verifique se ele atende às regras:

 O padrão esperado é que os nomes comecem com uma letra maiúscula e contenham apenas letras (sem números ou caracteres especiais).


```python
import re

usuario = input("Digite o seu usuário: ")

if re.fullmatch(r"[A-Z][a-zA-Z]*", usuario):
    print("Usuário válido!")
else:
    print("Usuário inválido!")
```

# Verificando o formato de um CPF

Um programa que solicite o CPF de um cliente e verifica se ele está no formato correto.

O formato esperado do CPF é: três blocos de 3 dígitos separados por pontos (.), seguidos por um bloco de 2 dígitos separados por um traço (-).

```python
import re

cpf = input("Digite o CPF: ")

if re.fullmatch(r"\d{3}\.\d{3}\.\d{3}-\d{2}", cpf):
    print("CPF no formato válido!")
else:
    print("CPF em formato inválido!")
```
